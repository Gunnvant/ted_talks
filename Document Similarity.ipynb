```python
import os
import pandas as pd
base_dir="E:\\Kaggle\\ted-talks"
os.chdir(base_dir)
transcripts=pd.read_csv("transcripts.csv")
```

```python
from sklearn.feature_extraction import text
```

```python
transcripts.head()
```

```python
transcripts_list=transcripts['transcript'].tolist()
```

```python
len(transcripts_list)
```

```python
transcripts.shape[0]
```

```python
tfidf=text.TfidfVectorizer(input=transcripts_list,stop_words="english")
```

```python
tfidf_matrix=tfidf.fit_transform(transcripts_list)
```

```python
tfidf_matrix.shape
```

```python
tfidf_df=pd.DataFrame({"token":tfidf.get_feature_names(),"tfidf":tfidf.idf_})
```

```python
tfidf_df.head()
```

```python
tfidf_df.sort_values("tfidf",ascending=False).head()
```
